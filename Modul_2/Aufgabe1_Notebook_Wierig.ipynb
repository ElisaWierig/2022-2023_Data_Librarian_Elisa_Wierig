```python
import json
data1 = json.load(open("archiv.json"))
data2 = json.load(open("museum.json"))
```

```python
import pandas as pd
```

```python
data1.keys()
```

```python
df_search1 = pd.DataFrame(data1["items"])
df_search2 = pd.DataFrame(data2["items"])
```

```python
df_search1
```

```python
df_search2
```

```python
df_both = pd.concat([df_search1, df_search2])
```

```python
df_both
```

```python
df_both.columns
```

```python
df_both.url
```

```python
df_both[["label", "author", "year", "type", "url"]]
```

```python
df_both_selection = df_both[["label", "author", "year", "type", "url"]]
```

```python
df_both_selection
```

```python
df_both_selection = df_both_selection.loc[df_both_selection["author"].isnull() == False]
```

```python
df_both_selection
```

```python
df_both_selection.loc[df_both_selection.type == "Publication"]
```

```python
df_both_selection = df_both_selection.drop_duplicates(subset = "label", keep = "first")
```

```python
df_both_selection
```

```python
df_both_selection["year"].unique()
```

```python
df_both_selection["year_correct"] = df_both_selection.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
df_both_selection = df_both_selection.loc[df_both_selection.year_correct == True]
```

```python
df_both_selection
```

```python
df_both_selection_no_nan = df_both_selection.dropna()
```

```python
df_both_selection_no_nan
```

```python
df_both_selection = df_both_selection.rename(columns={"label": "title", "author": "authors", "url": "link"})
df_both_selection
```

```python
df_both_selection["collected_by"] = "Elisa"
df_both_selection
```

```python
df_both_selection["id"] = df_both_selection.index
```

```python
df_both_selection["number_authors"] = df_both_selection.apply( lambda row: len(row.authors), axis=1)
df_both_selection
```

```python
df_both_selection.to_json("Aufgabe1_output.json", orient="records")
```
